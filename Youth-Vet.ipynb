```python
import pandas as pd
```

```python
neww = pd.read_csv("new_stats.csv")
new3 = neww["Team"].unique()
new3.sort()
```

```python
youth = pd.DataFrame(columns = ["Team", "Youth", "Vets"])
grouped = neww.groupby("Team")

for name,group in grouped:
    team_groups = pd.DataFrame({"EXP" : grouped.get_group(name)["EXP"]})
    
    df2 = pd.DataFrame({"Team" : name,
                       "Youth" : team_groups[team_groups.EXP < 5].count(),
                       "Vets" : team_groups[team_groups.EXP >= 5].count()})
    
    youth = pd.concat([youth, df2])
```

```python
youth.sort_values("Vets", ascending = False, inplace = True)
output = None
```

```python
youth["Mindset"] = ""

youth.loc[(youth["Youth"] >= 4), "Mindset"] = "Rebuilding"
youth.loc[(youth["Youth"] <= 3), "Mindset"] = "Win Now"

youth.reset_index(drop = True, inplace = True)
youth.style.set_properties(**{'text-align': 'center'})
```

```python
youth.to_csv("finalYouthVet.csv")
```

Teams that 5+ vets in their top seven players are generally in a win now mindset. Some exceptions are given to more youth oriented teams that have exceptional talents like Denver. This doesn't account for rookies since we don't know how much a team plans to incorporate that player in his first season. 

Teams that have just drafted in the top 5 in the lottery usually plan to have that player have a bigger role immediately. So teams like the Pelicans, Grizzlies, and Knicks would look a little younger than their rank right now.
