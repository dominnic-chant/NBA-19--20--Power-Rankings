```python
import pandas as pd
```

Uses both offense notebook and draymond notebook

```python
offense = pd.read_csv("top_offense.csv")
offense = offense.iloc[:,1:]
```

```python
defense = pd.read_csv("top_defense.csv")
defense = defense.iloc[:,1:]
```

```python
top_stars = pd.merge(offense, defense, how = "outer")
# top_stars = offense.append(defense, sort = False)
top_stars["dvalue"].fillna(0, inplace = True)
top_stars["ovalue"].fillna(0, inplace = True)
top_stars["value"] = top_stars["ovalue"] + top_stars["dvalue"]
top_stars.sort_values(["value", "ovalue"], inplace = True, ascending = False)
```

```python
top_stars= top_stars[["Name", "Team", "ovalue", "dvalue", "value"]]
top_stars.reset_index(drop = True, inplace = True)
top_stars
```

```python
top_stars_counts = pd.DataFrame({"Count" : top_stars["Team"].value_counts()})
top_stars_counts.reset_index(inplace = True)
top_stars_counts.rename(columns = {"index" : "Team"}, inplace = True)
top_stars_counts

top_stars_counts["Grade"] = ""
top_stars_counts.loc[(top_stars_counts["Count"] == 4), "Grade"] = "A"
top_stars_counts.loc[(top_stars_counts["Count"] == 3), "Grade"] = "B"
top_stars_counts.loc[(top_stars_counts["Count"] == 2), "Grade"] = "C"
top_stars_counts.loc[(top_stars_counts["Count"] == 1), "Grade"] = "D"
top_stars_counts.loc[(top_stars_counts["Count"] == 0), "Grade"] = "F"

top_stars_counts.style.set_properties(**{'text-align': 'center'})
```

```python
top_stars_counts.to_csv("finalStars.csv")
```
