```python
import pandas as pd
```

### Offense

```python
def offense():
    df = pd.read_csv("new_stats.csv")

    pts = pd.DataFrame({"True Shooting": round(df.groupby("Team")["TS%"].mean(), 2),
                        "PTS/Team" : round(df.groupby("Team")["PTS/G"].sum(), 2)})

    pts.reset_index(inplace = True)
    pts["PTSRank"] = pts["PTS/Team"].rank().astype(int)
    pts["TrueShootingRank"] = pts["True Shooting"].rank().astype(int)
    
    
    three_point_attempt = pd.DataFrame({"3PA" : df.groupby("Team")["3PA"].sum()})
    three_point_attempt.reset_index(inplace = True)

    three_point_percent = pd.DataFrame({"3P%" : round(df.groupby("Team")["3P%"].mean(), 2)})
    three_point_percent.reset_index(inplace = True)
    
    three_point_attempt.sort_values("3PA", ascending = False, inplace = True)
    three_point_percent.sort_values("3P%", ascending = False, inplace = True)

    three_point_attempt["3PARank"] = three_point_attempt["3PA"].rank().astype(int)
    three_point_percent["3P%Rank"] = three_point_percent["3P%"].rank().astype(int)

    
    three_point = pd.merge(three_point_attempt, three_point_percent, on = "Team")   #merge columns
    three_point = three_point[["Team", "3PA", "3P%", "3PARank", "3P%Rank"]]   #reordering columns
    
    
    team_offense = pd.merge(three_point, pts, on = "Team")
    team_offense = team_offense[["Team", "3PA", "3P%", "PTS/Team", "True Shooting", "3PARank", "3P%Rank", "PTSRank", "TrueShootingRank"]]
    
    
    team_offense["TeamRank"] = team_offense["3PARank"] + team_offense["3P%Rank"] + team_offense["PTSRank"] + team_offense["TrueShootingRank"]
    team_offense.sort_values(["TeamRank", "TrueShootingRank", "PTSRank"], ascending = False, inplace = True)
    team_offense.reset_index(drop = True, inplace = True)
    
    
    team_offense.loc[(team_offense["TeamRank"] <= 121.9) & (team_offense["TeamRank"] >= 114), ["Grade", 'int_grade']] = ['A+', 12]
    team_offense.loc[(team_offense["TeamRank"] <= 113.9) & (team_offense["TeamRank"] >= 106), ["Grade", 'int_grade']] = ['A',11]
    team_offense.loc[(team_offense["TeamRank"] <= 105.9) & (team_offense["TeamRank"] >= 98), ["Grade", 'int_grade']] = ['A-', 10]
    team_offense.loc[(team_offense["TeamRank"] <= 97.9) & (team_offense["TeamRank"] >= 90), ["Grade", 'int_grade']] = ['B+', 9]
    team_offense.loc[(team_offense["TeamRank"] <= 89.9) & (team_offense["TeamRank"] >= 82), ["Grade", 'int_grade']] = ['B', 8]
    team_offense.loc[(team_offense["TeamRank"] <= 81.9) & (team_offense["TeamRank"] >= 74), ["Grade", 'int_grade']] = ['B-', 7]
    team_offense.loc[(team_offense["TeamRank"] <= 73.9) & (team_offense["TeamRank"] >= 66), ["Grade", 'int_grade']] = ['C+', 6]
    team_offense.loc[(team_offense["TeamRank"] <= 65.9) & (team_offense["TeamRank"] >= 58), ["Grade", 'int_grade']] = ['C', 5]
    team_offense.loc[(team_offense["TeamRank"] <= 57.9) & (team_offense["TeamRank"] >= 50), ["Grade", 'int_grade']] = ['C-', 4]
    team_offense.loc[(team_offense["TeamRank"] <= 49.9) & (team_offense["TeamRank"] >= 42), ["Grade", 'int_grade']] = ['D+', 3]
    team_offense.loc[(team_offense["TeamRank"] <= 41.9) & (team_offense["TeamRank"] >= 34), ["Grade", 'int_grade']] = ['D', 2]
    team_offense.loc[(team_offense["TeamRank"] <= 33.9) & (team_offense["TeamRank"] >= 26), ["Grade", 'int_grade']] = ['D-', 1]
    team_offense.loc[(team_offense["TeamRank"] <= 25.9) & (team_offense["TeamRank"] >= 18), ["Grade", 'int_grade']] = ['F', 0]
    
    return team_offense
```

### Defense

```python
def defense():
    df = pd.read_csv("new_stats.csv") 
    
    
    drayavg = pd.DataFrame({"DRAYavg" : round(df.groupby("Team")["DRAYMOND"].mean(), 2)})
    possavg = pd.DataFrame({"POSSavg" : round(df.groupby("Team")["DPOSS"].sum(), 2)})
    
    drayavg.reset_index(inplace = True)
    possavg.reset_index(inplace = True)

    drayavg.sort_values("DRAYavg", ascending = False, inplace = True)
    possavg.sort_values("POSSavg", ascending = False, inplace = True)

    drayavg["DRAYrank"] = drayavg["DRAYavg"].rank().astype(int)
    possavg["POSSrank"] = possavg["POSSavg"].rank().astype(int)

    team_defense = pd.merge(drayavg, possavg, on = "Team")
    team_defense = team_defense[["Team", "DRAYavg", "POSSavg", "DRAYrank", "POSSrank"]]
    
    team_defense["TeamRank"] = team_defense["DRAYrank"] + team_defense["POSSrank"]
    team_defense.sort_values("TeamRank", ascending = False, inplace = True)
    

    team_defense.loc[(team_defense["TeamRank"] <= 65.4) & (team_defense["TeamRank"] >= 61), ["Grade", 'int_grade']] = ['A+', 12]
    team_defense.loc[(team_defense["TeamRank"] <= 60.9) & (team_defense["TeamRank"] >= 56.5), ["Grade", 'int_grade']] = ['A', 11]
    team_defense.loc[(team_defense["TeamRank"] <= 56.4) & (team_defense["TeamRank"] >= 52), ["Grade", 'int_grade']] = ['A-', 10]
    team_defense.loc[(team_defense["TeamRank"] <= 51.9) & (team_defense["TeamRank"] >= 47.5), ["Grade", 'int_grade']] = ['B+', 9]
    team_defense.loc[(team_defense["TeamRank"] <= 47.4) & (team_defense["TeamRank"] >= 43), ["Grade", 'int_grade']] = ['B', 8]
    team_defense.loc[(team_defense["TeamRank"] <= 42.9) & (team_defense["TeamRank"] >= 38.5), ["Grade", 'int_grade']] = ['B-', 7]
    team_defense.loc[(team_defense["TeamRank"] <= 38.4) & (team_defense["TeamRank"] >= 34), ["Grade", 'int_grade']] = ['C+', 6]
    team_defense.loc[(team_defense["TeamRank"] <= 33.9) & (team_defense["TeamRank"] >= 29.5), ["Grade", 'int_grade']] = ['C', 5]
    team_defense.loc[(team_defense["TeamRank"] <= 29.4) & (team_defense["TeamRank"] >= 25), ["Grade", 'int_grade']] = ['C-', 4]
    team_defense.loc[(team_defense["TeamRank"] <= 24.9) & (team_defense["TeamRank"] >= 20.5), ["Grade", 'int_grade']] = ['D+', 3]
    team_defense.loc[(team_defense["TeamRank"] <= 20.4) & (team_defense["TeamRank"] >= 16), ["Grade", 'int_grade']] = ['D', 2]
    team_defense.loc[(team_defense["TeamRank"] <= 15.9) & (team_defense["TeamRank"] >= 11.5), ["Grade", 'int_grade']] = ['D-', 1]
    team_defense.loc[(team_defense["TeamRank"] <= 11.4) & (team_defense["TeamRank"] >= 7), ["Grade", 'int_grade']] = ['F', 0]

    team_defense.reset_index(drop = True, inplace = True)
    
    return team_defense
```

### Clutch

```python
def clutch():
    clutch = pd.read_csv("clutch.csv")
    clutch = clutch.iloc[:,1:13]
    
    clutch1 = clutch[(clutch["Cltch"] > 99) & (clutch["Cltch.1"] >= "51%")].sort_values("Cltch.1", ascending = False)
    clutch1["Score_1"] = clutch1["Cltch.1"].rank().astype(int)
    clutch1 = clutch1[["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1", "Score_1"]]
    
    clutch2 = clutch[(clutch["Cltch2"] > 12) & (clutch["Cltch2.1"] > "50.5%")].sort_values("Cltch2.1", ascending = False)
    clutch2["Score_2"] = clutch2["Cltch2.1"].rank().astype(int)
    clutch2["Score_2"] = (clutch2["Score_2"] * 1.3)
    clutch2 = clutch2[["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1", "Score_2"]]
    
    most_clutch = pd.merge(clutch1, clutch2, on = ["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1"], how = "outer")
    most_clutch.fillna(0, inplace = True)
    most_clutch["Score"] = (most_clutch["Score_1"] + most_clutch["Score_2"])
    
    most_clutch.sort_values(["Score", "Score_2"], ascending = False, inplace = True)
    most_clutch.reset_index(drop = True, inplace = True)

    most_clutch["Team"].value_counts()

    most_clutch_counts = pd.DataFrame({"Count" : most_clutch["Team"].value_counts()})
    most_clutch_counts.reset_index(inplace = True)
    most_clutch_counts.rename(columns = {"index" : "Team"}, inplace = True)
    
    return most_clutch_counts
```

### Stars

```python
def stars():
    offense = pd.read_csv("new_stats.csv")
    top_offense = (offense.loc[(offense["TS%"] > 0.525) & ((offense["3PA"] + offense["2PA"])>= 16) & (offense["PTS/G"] >= 20.5), ["Name", "Team", "3PA", "2PA", "PTS/G", "TS%"]]
                          .sort_values("TS%", ascending = False))
    top_offense["orank"] = top_offense["TS%"].rank().astype(int)

    dfo = top_offense

    
    defense = pd.read_csv("new_stats.csv")
    top_defense = (defense.loc[(defense["DPOSS"] >= 4450) & (defense["DRAYMOND"] > 0.7), ["Name", "Team", "DPOSS", "DRAYMOND"]]
                          .sort_values("DRAYMOND", ascending = False))
    top_defense["drank"] = top_defense["DRAYMOND"].rank().astype(int)

    dfd = top_defense


    top_stars = pd.merge(dfo, dfd, how = "outer")
    top_stars["drank"].fillna(0, inplace = True)
    top_stars["orank"].fillna(0, inplace = True)
    
    top_stars = top_stars[["Name", "Team", "orank", "drank"]]
    top_stars.reset_index(drop = True, inplace = True)
    top_stars = top_stars.iloc[:55, :]
    
    
    top_stars['Value'] = 5
    top_stars.loc[((top_stars['orank'] + top_stars['drank']) / 2 <= 29.0), 'Value'] = 4
    top_stars.loc[((top_stars['orank'] + top_stars['drank']) / 2 <= 22.0), 'Value'] = 3
    top_stars.loc[((top_stars['orank'] + top_stars['drank']) / 2 <= 15.0), 'Value'] = 2
    top_stars.loc[((top_stars['orank'] + top_stars['drank']) / 2 <= 8.0), 'Value'] = 1
    
    top_stars.sort_values('Value', ascending = False, inplace = True)
    
    star_value = pd.DataFrame({'Star Value' : top_stars.groupby("Team")["Value"].sum()})
    star_value.reset_index(inplace = True)
    
    star_value.loc[(star_value["Star Value"] <= 10), ["Grade", 'int_grade']] = ['A', 12]
    star_value.loc[(star_value["Star Value"] <= 6), ["Grade", 'int_grade']] = ['B', 9]
    star_value.loc[(star_value["Star Value"] <= 3), ["Grade", 'int_grade']] = ['C', 6]
    star_value.loc[(star_value["Star Value"] <= 2), ["Grade", 'int_grade']] = ['D', 3]
    star_value.loc[(star_value["Star Value"] <= 1), ["Grade", 'int_grade']] = ['F', 1]
    
    star_value.sort_values('Star Value', ascending = False, inplace = True)
    
    return star_value
```

### Youth-Vet

```python
def youthVet():
    df = pd.read_csv("new_stats.csv")
    
    youth = pd.DataFrame(columns = ["Team", "Youth", "Vets"])
    grouped = df.groupby("Team")

    for name,group in grouped:
        team_groups = pd.DataFrame({"EXP" : grouped.get_group(name)["EXP"]})

        df2 = pd.DataFrame({"Team" : name,
                           "Youth" : team_groups[team_groups.EXP < 5].count(),
                           "Vets" : team_groups[team_groups.EXP >= 5].count()})

        youth = pd.concat([youth, df2])
        
    youth.sort_values("Vets", ascending = False, inplace = True)
    
    youth["Mindset"] = ""

    youth.loc[(youth["Youth"] >= 4), "Mindset"] = "Rebuilding"
    youth.loc[(youth["Youth"] <= 3), "Mindset"] = "Win Now"

    youth.reset_index(drop = True, inplace = True)
    
    
    return youth
```

### Final

```python
def main():
    odf = offense()
    ddf = defense()
    cdf = clutch()
    sdf = stars()
    ydf = youthVet()
    
    final = pd.merge(odf[['Team', 'Grade', 'int_grade']], ddf[['Team', 'Grade', 'int_grade']], on = "Team")
    final = pd.merge(final, sdf[['Team', 'Grade', 'int_grade']], on = "Team", how = "left")
    final = pd.merge(final, cdf[['Team', 'Count']], on = "Team", how = "left")
    final = pd.merge(final, ydf[['Team', 'Mindset']], on = "Team")
    final.rename(columns = {"Grade_x" : "Offense", "int_grade_x" : "int_Offense",
                            "Grade_y" : "Defense", "int_grade_y" : "int_Defense",
                            "Grade" : "Star Value", "int_grade" : "int_Star",
                            "Count" : "Clutch Player Count"}, 
                            inplace = True)
    
    final[['int_Star', 'Clutch Player Count']] = final[['int_Star', 'Clutch Player Count']].fillna(0)
    final["Clutch Player Count"] = final["Clutch Player Count"].astype(int)
    final[['Star Value']] = final[['Star Value']].fillna('F')
    
    final['Overall'] = final['int_Offense'] + final['int_Defense'] + (final['int_Star'] * 0.25)
    
    final = final.loc[:, ['Team', 'Offense', 'Defense', 'Star Value', 'Clutch Player Count', 'Overall', 'Mindset']]

    final.sort_values(['Overall', 'Clutch Player Count'], ascending = False, inplace = True)
    final.reset_index(drop = True, inplace = True)
    return(final)
```

```python
main()
```
