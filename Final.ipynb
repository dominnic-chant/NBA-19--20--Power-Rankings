```python
import pandas as pd
```

```python
offense = pd.read_csv("finalOffense.csv")
defense = pd.read_csv("finalDefense.csv")
clutch = pd.read_csv("finalClutch.csv")
stars = pd.read_csv("finalStars.csv")
youthvet = pd.read_csv("finalYouthVet.csv")
```

```python
offense1 = offense[["Team","Grade"]]
defense1 = defense[["Team","Grade"]]
stars1 = stars[["Team", "Grade"]]
clutch1 = clutch[["Team", "Count"]]
youthvet1 = youthvet[["Team", "Mindset"]]
```

```python
conclusion = pd.merge(offense1, defense1, on = "Team")
conclusion1 = pd.merge(conclusion, stars1, on = "Team", how = "left").fillna("F")
conclusion2 = pd.merge(conclusion1, clutch1, on = "Team", how = "left").fillna(0)
conclusion3 = pd.merge(conclusion2, youthvet1, on = "Team")
conclusion3["Count"] = conclusion3["Count"].astype(int)
conclusion3.rename(columns = {"Grade_x" : "Offense",
                               "Grade_y" : "Defense",
                               "Grade" : "Star Count",
                               "Count" : "Clutch Player Count"}, inplace = True)

conclusion3.style.set_properties(**{'text-align': 'center'})
```
