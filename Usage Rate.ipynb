```python
import pandas as pd
```

```python
neww = pd.read_csv("new_stats.csv")
new3 = neww["Team"].unique()
new3.sort()

usage = pd.DataFrame({"Team" : new3,
                      "Avg Usage" : round(neww.groupby("Team")["USG%"].mean(), 1),
                      "Max Usage" : neww.groupby("Team")["USG%"].max()})
usage.reset_index(drop = True, inplace = True)
```

```python
usage.sort_values("Max Usage", ascending = False)
```
