## Team Defense

```python
new3 = neww["Team"].unique()
new3.sort()

drayavg = pd.DataFrame({"Team": new3,
                        "DRAYavg" : round(neww.groupby("Team")["DRAYMOND"].mean(), 2)})

possavg = pd.DataFrame({"Team" : new3,
                       "POSSavg" : round(neww.groupby("Team")["DPOSS"].sum(), 2)})
```

```python
drayavg.reset_index(drop = True, inplace = True)
possavg.reset_index(drop = True, inplace = True)

drayavg.sort_values("DRAYavg", ascending = False, inplace = True)
possavg.sort_values("POSSavg", ascending = False, inplace = True)

drayavg["DRAYvalue"] = drayavg["DRAYavg"].rank().astype(int)
possavg["POSSvalue"] = possavg["POSSavg"].rank().astype(int)

team_defense = pd.merge(drayavg, possavg, on = "Team")
```

```python
team_defense = team_defense[["Team", "DRAYavg", "POSSavg", "DRAYvalue", "POSSvalue"]]
```

```python
team_defense["TeamRank"] = team_defense["DRAYvalue"] + team_defense["POSSvalue"]
team_defense.sort_values("TeamRank", ascending = False, inplace = True)
```

```python
team_defense["Grade"] = ""

team_defense.loc[(team_defense["TeamRank"] <= 65.4) & (team_defense["TeamRank"] >= 61), "Grade"] = "A+"
team_defense.loc[(team_defense["TeamRank"] <= 60.9) & (team_defense["TeamRank"] >= 56.5), "Grade"] = "A"
team_defense.loc[(team_defense["TeamRank"] <= 56.4) & (team_defense["TeamRank"] >= 52), "Grade"] = "A-"
team_defense.loc[(team_defense["TeamRank"] <= 51.9) & (team_defense["TeamRank"] >= 47.5), "Grade"] = "B+"
team_defense.loc[(team_defense["TeamRank"] <= 47.4) & (team_defense["TeamRank"] >= 43), "Grade"] = "B"
team_defense.loc[(team_defense["TeamRank"] <= 42.9) & (team_defense["TeamRank"] >= 38.5), "Grade"] = "B-"
team_defense.loc[(team_defense["TeamRank"] <= 38.4) & (team_defense["TeamRank"] >= 34), "Grade"] = "C+"
team_defense.loc[(team_defense["TeamRank"] <= 33.9) & (team_defense["TeamRank"] >= 29.5), "Grade"] = "C"
team_defense.loc[(team_defense["TeamRank"] <= 29.4) & (team_defense["TeamRank"] >= 25), "Grade"] = "C-"
team_defense.loc[(team_defense["TeamRank"] <= 24.9) & (team_defense["TeamRank"] >= 20.5), "Grade"] = "D+"
team_defense.loc[(team_defense["TeamRank"] <= 20.4) & (team_defense["TeamRank"] >= 16), "Grade"] = "D"
team_defense.loc[(team_defense["TeamRank"] <= 15.9) & (team_defense["TeamRank"] >= 11.5), "Grade"] = "D-"
team_defense.loc[(team_defense["TeamRank"] <= 11.4) & (team_defense["TeamRank"] >= 7), "Grade"] = "F"

team_defense.reset_index(drop = True, inplace = True)
team_defense.style.set_properties(**{'text-align': 'center'})
```

```python
team_defense.to_csv("finalDefense.csv")
```
