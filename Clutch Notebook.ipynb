```python
import pandas as pd
clutch = pd.read_csv("clutch.csv")
```

```python
clutch = clutch.iloc[:,1:13]
```

```python
clutch1 = clutch[(clutch["Cltch"] > 99) & (clutch["Cltch.1"] >= "51%")].sort_values("Cltch.1", ascending = False)
clutch1["Score_1"] = clutch1["Cltch.1"].rank().astype(int)
clutch1 = clutch1[["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1", "Score_1"]]
clutch1
```

```python
clutch2 = clutch[(clutch["Cltch2"] > 12) & (clutch["Cltch2.1"] > "50.5%")].sort_values("Cltch2.1", ascending = False)
clutch2["Score_2"] = clutch2["Cltch2.1"].rank().astype(int)
clutch2["Score_2"] = (clutch2["Score_2"] * 1.3)
clutch2 = clutch2[["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1", "Score_2"]]
clutch2
```

```python
most_clutch = pd.merge(clutch1, clutch2, on = ["Player", "Pos", "Team", "Cltch", "Cltch2", "Cltch.1", "Cltch2.1"], how = "outer")
most_clutch.fillna(0, inplace = True)
most_clutch["Score"] = (most_clutch["Score_1"] + most_clutch["Score_2"])
```

```python
most_clutch.sort_values(["Score", "Score_2"], ascending = False, inplace = True)
most_clutch.reset_index(drop = True, inplace = True)
most_clutch
```

```python
most_clutch["Team"].value_counts()

most_clutch_counts = pd.DataFrame({"Count" : most_clutch["Team"].value_counts()})
most_clutch_counts.reset_index(inplace = True)
most_clutch_counts.rename(columns = {"index" : "Team"}, inplace = True)
most_clutch_counts
```

```python
most_clutch_counts.to_csv("finalClutch.csv")
```

Top clutch players ranked by combined score of semi-clutch and clutch percentages with clutch given extra value. Tie breakers are settled by higher clutch percentage.
