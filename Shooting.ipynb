```python
import pandas as pd
```

## Sum of Average Points Per Game and Player

```python
neww = pd.read_csv("new_stats.csv")
new3 = neww["Team"].unique()
new3.sort()

pts = pd.DataFrame({"Team" : new3,
                   "True Shooting": round(neww.groupby("Team")["TS%"].mean(), 2),
                   "PTS/Team" : round(neww.groupby("Team")["PTS/G"].sum(), 2)})
pts.reset_index(drop = True, inplace = True)
```

```python
pts["PTSRank"] = pts["PTS/Team"].rank().astype(int)
pts["TrueShootingRank"] = pts["True Shooting"].rank().astype(int)
pts.sort_values(["PTSRank","TrueShootingRank"], ascending = False)
```

Average points and average true shooting % by team 

## Shoots the Most Threes and Effieciently 

```python
three_point_attempt = pd.DataFrame({"Team" : new3,
                                   "3PA" : neww.groupby("Team")["3PA"].sum()})
three_point_attempt.reset_index(drop = True, inplace = True)

three_point_percent = pd.DataFrame({"Team" : new3,
                                   "3P%" : round(neww.groupby("Team")["3P%"].mean(), 2)})
three_point_percent.reset_index(drop = True, inplace = True)
```

```python
three_point_attempt.sort_values("3PA", ascending = False, inplace = True)
three_point_percent.sort_values("3P%", ascending = False, inplace = True)

three_point_attempt["3PARank"] = three_point_attempt["3PA"].rank().astype(int)
three_point_percent["3P%Rank"] = three_point_percent["3P%"].rank().astype(int)
```

```python
three_point = pd.merge(three_point_attempt, three_point_percent, on = "Team")   #merge columns
three_point = three_point[["Team", "3PA", "3P%", "3PARank", "3P%Rank"]]   #reordering columns
```

```python
three_point
```

Teams ranked in order based on a combined score from best 3 point percentage and most 3 point attempts.

All ties are settled by the better 3 point percentage then 3 point attempts.

```python
team_offense = pd.merge(three_point, pts, on = "Team")
team_offense = team_offense[["Team", "3PA", "3P%", "PTS/Team", "True Shooting", "3PARank", "3P%Rank", "PTSRank", "TrueShootingRank"]]
```

```python
team_offense["TeamRank"] = team_offense["3PARank"] + team_offense["3P%Rank"] + team_offense["PTSRank"] + team_offense["TrueShootingRank"]
team_offense.sort_values(["TeamRank", "TrueShootingRank", "PTSRank"], ascending = False, inplace = True)
team_offense.reset_index(drop = True, inplace = True)
team_offense
```

```python
team_offense["Grade"] = ""
team_offense.loc[(team_offense["TeamRank"] <= 121.9) & (team_offense["TeamRank"] >= 114), "Grade"] = "A+"
team_offense.loc[(team_offense["TeamRank"] <= 113.9) & (team_offense["TeamRank"] >= 106), "Grade"] = "A"
team_offense.loc[(team_offense["TeamRank"] <= 105.9) & (team_offense["TeamRank"] >= 98), "Grade"] = "A-"
team_offense.loc[(team_offense["TeamRank"] <= 97.9) & (team_offense["TeamRank"] >= 90), "Grade"] = "B+"
team_offense.loc[(team_offense["TeamRank"] <= 89.9) & (team_offense["TeamRank"] >= 82), "Grade"] = "B"
team_offense.loc[(team_offense["TeamRank"] <= 81.9) & (team_offense["TeamRank"] >= 74), "Grade"] = "B-"
team_offense.loc[(team_offense["TeamRank"] <= 73.9) & (team_offense["TeamRank"] >= 66), "Grade"] = "C+"
team_offense.loc[(team_offense["TeamRank"] <= 65.9) & (team_offense["TeamRank"] >= 58), "Grade"] = "C"
team_offense.loc[(team_offense["TeamRank"] <= 57.9) & (team_offense["TeamRank"] >= 50), "Grade"] = "C-"
team_offense.loc[(team_offense["TeamRank"] <= 49.9) & (team_offense["TeamRank"] >= 42), "Grade"] = "D+"
team_offense.loc[(team_offense["TeamRank"] <= 41.9) & (team_offense["TeamRank"] >= 34), "Grade"] = "D"
team_offense.loc[(team_offense["TeamRank"] <= 33.9) & (team_offense["TeamRank"] >= 26), "Grade"] = "D-"
team_offense.loc[(team_offense["TeamRank"] <= 25.9) & (team_offense["TeamRank"] >= 18), "Grade"] = "F"

team_offense.style.set_properties(**{'text-align': 'center'})
```
