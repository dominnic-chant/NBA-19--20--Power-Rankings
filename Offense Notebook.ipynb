```python
import pandas as pd
offense = pd.read_csv("new_stats.csv")
top_offense = offense[(offense["TS%"] > 0.525) & ((offense["3PA"] + offense["2PA"])>= 16) & (offense["PTS/G"] >= 20.5)].sort_values("TS%", ascending = False)
top_offense = top_offense.loc[:,["Name", "Team", "3PA", "2PA", "PTS/G", "TS%"]]

```

```python
top_offense["ovalue"] = top_offense["TS%"].rank().astype(int)
top_offense
```
